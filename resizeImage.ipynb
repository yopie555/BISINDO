```python
from PIL import Image  
import os  

# Fungsi untuk meresize semua gambar dalam folder  
def resize_images_in_folder(input_directory, output_directory):  
    # Membuat direktori output jika belum ada  
    os.makedirs(output_directory, exist_ok=True)  
    
    # Mendapatkan semua file gambar dalam folder input  
    for count, filename in enumerate(os.listdir(input_directory), start=1):  
        # Mengambil setiap file dengan ekstensi gambar  
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):  
            image_path = os.path.join(input_directory, filename)  
            with Image.open(image_path) as img:  
                # Resize gambar menjadi 640x640  
                img = img.resize((224, 224))  
                # Menyimpan gambar yang telah di-resize  
                img.save(os.path.join(output_directory, f'YellowMosaic{count:03d}.jpg'))  
                print(f'Saved: YellowMosaic{count:03d}.jpg')  

input_dir = '/Users/nero555/Documents/Dev/CV/kedelai-DS/dataset/original/YellowMosaic'  # Path gambar yang akan di-resize
output_dir = '/Users/nero555/Documents/Dev/CV/kedelai-DS/dataset/resize/YellowMosaic'  # Direktori output

resize_images_in_folder(input_dir, output_dir)  # Resize semua gambar di folder
# Contoh penggunaan  
```

```python
from PIL import Image  
import os  

# Fungsi untuk augmentasi gambar  
def augment_images_in_folder(input_directory, output_directory):  
    # Membuat direktori output jika belum ada  
    os.makedirs(output_directory, exist_ok=True)  

    # Sudut rotasi yang akan digunakan  
    rotation_angles = [20, 45, 90]  

    # Mendapatkan semua file gambar dalam folder input  
    for filename in os.listdir(input_directory):  
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):  
            image_path = os.path.join(input_directory, filename)  

            # Membaca gambar asli  
            with Image.open(image_path) as img:  
                # Melakukan rotasi dan flipping gambar  
                for angle in rotation_angles:  
                    # Rotasi gambar  
                    rotated_img = img.rotate(angle)  
                    # Menyimpan gambar yang telah dirotasi  
                    rotated_img.save(os.path.join(output_directory, f'{filename[:-4]}_rotated_{angle}.jpg'))  

                    # Flipping gambar yang telah dirotasi secara horizontal  
                    flipped_horizontal = rotated_img.transpose(method=Image.FLIP_LEFT_RIGHT)  
                    flipped_horizontal.save(os.path.join(output_directory, f'{filename[:-4]}_rotated_{angle}_flipped_horizontal.jpg'))  

                    # Flipping gambar yang telah dirotasi secara vertikal  
                    flipped_vertical = rotated_img.transpose(method=Image.FLIP_TOP_BOTTOM)  
                    flipped_vertical.save(os.path.join(output_directory, f'{filename[:-4]}_rotated_{angle}_flipped_vertical.jpg'))  

                print(f'Processed: {filename}')  

# Contoh penggunaan  
input_dir = '/Users/nero555/Documents/Data Science/Penelitian/resize/BacterialPustule'  # Path gambar yang akan di-resize
output_dir = '/Users/nero555/Documents/Data Science/Penelitian/augmentasi/BacterialPustule'  # Direktori output

augment_images_in_folder(input_dir, output_dir)  # Melakukan augmentasi pada semua gambar di folder
```

```python
from PIL import Image, ImageEnhance, ImageFilter  
import numpy as np  
import os  
import random  

# Fungsi untuk augmentasi gambar  
def augment_images_in_folder(input_directory, output_directory):  
    # Membuat direktori output jika belum ada  
    os.makedirs(output_directory, exist_ok=True)  

    # Sudut rotasi yang akan digunakan  
    rotation_angles = [20, 45, 90]  

    # Mendapatkan semua file gambar dalam folder input  
    for filename in os.listdir(input_directory):  
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):  
            image_path = os.path.join(input_directory, filename)  

            # Membaca gambar asli  
            with Image.open(image_path) as img:  
                # Melakukan rotasi dan flipping gambar  
                for angle in rotation_angles:  
                    # Rotasi gambar  
                    rotated_img = img.rotate(angle)  
                    # Menyimpan gambar yang telah dirotasi  
                    rotated_img.save(os.path.join(output_directory, f'{filename[:-4]}_rotated_{angle}.jpg'))  

                    # Flipping gambar yang telah dirotasi secara horizontal  
                    flipped_horizontal = rotated_img.transpose(method=Image.FLIP_LEFT_RIGHT)  
                    flipped_horizontal.save(os.path.join(output_directory, f'{filename[:-4]}_rotated_{angle}_flipped_horizontal.jpg'))  

                    # Flipping gambar yang telah dirotasi secara vertikal  
                    flipped_vertical = rotated_img.transpose(method=Image.FLIP_TOP_BOTTOM)  
                    flipped_vertical.save(os.path.join(output_directory, f'{filename[:-4]}_rotated_{angle}_flipped_vertical.jpg'))  

                    # Augmentasi Gaussian Blur  
                    blurred_img = rotated_img.filter(ImageFilter.GaussianBlur(radius=2))  # Mengatur radius sesuai kebutuhan  
                    blurred_img.save(os.path.join(output_directory, f'{filename[:-4]}_rotated_{angle}_blurred.jpg'))  

                    # Augmentasi Random Brightness  
                    brightness_factor = random.uniform(0.5, 1.5)  # Mengatur nilai kecerahan acak antara 0.5 dan 1.5  
                    enhancer = ImageEnhance.Brightness(rotated_img)  
                    brightened_img = enhancer.enhance(brightness_factor)  
                    brightened_img.save(os.path.join(output_directory, f'{filename[:-4]}_rotated_{angle}_brightened.jpg'))  

                print(f'Processed: {filename}')  

# Contoh penggunaan  
input_dir = '/Users/nero555/Documents/Data Science/Penelitian/resize/YellowMosaic'  # Path gambar yang akan di-resize
output_dir = '/Users/nero555/Documents/Data Science/Penelitian/augmentasi/YellowMosaic'  # Direktori output

augment_images_in_folder(input_dir, output_dir)  # Melakukan augmentasi pada semua gambar di folder
```
